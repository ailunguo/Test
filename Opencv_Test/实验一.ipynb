```python
import numpy as np
import cv2 as cv
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

读入彩色图片

```python
region_img = cv.imread('/content/drive/MyDrive/zhihui/cat_2.jpg')
img = region_img.copy()
```

```python
plt.figure(figsize=(8, 6))
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.title('img')
plt.xticks([]),plt.yticks([])
plt.show()
```

```python
# 转换HSV空间
img_hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)

# 调亮
h, s, v = cv.split(img_hsv)
v_adjusted = cv.add(v, 50)
v_adjusted = np.clip(v_adjusted, 0, 255)
img_hsv_adjusted = cv.merge((h, s, v_adjusted))
bright_img = cv.cvtColor(img_hsv_adjusted, cv.COLOR_HSV2BGR)

plt.figure(figsize=(12, 10))
plt.subplot(121), plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB)), plt.title('Original')
plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(cv.cvtColor(bright_img, cv.COLOR_BGR2RGB)), plt.title('bright_image')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# 将图像转换为灰度图像
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.figure(figsize=(6, 4))
plt.imshow(cv.cvtColor(img_gray, cv.COLOR_BGR2RGB))
plt.show()
```

```python
# 绘制直方图
plt.hist(img_gray.ravel(), 256, [0,256]); plt.show()
```

```python
# 图像均衡化
equ = cv.equalizeHist(img_gray)
equ_img = np.hstack((img_gray, equ))
plt.figure(figsize=(10, 8))
plt.imshow(cv.cvtColor(equ_img, cv.COLOR_BGR2RGB)), plt.title(' Original--Equalization')
plt.show()
```
