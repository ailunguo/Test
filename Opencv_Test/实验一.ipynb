```python
import numpy as np
import cv2 as cv
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

读入彩色图片

```python
region_img = cv.imread('/content/drive/MyDrive/zhihui/cat_2.jpg')
img = region_img.copy()
```

```python
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.figure(figsize=(8, 6))
plt.imshow(img_rgb)
plt.title('img')
plt.xticks([]),plt.yticks([])
plt.show()
```

```python
# 转换HSV空间
img_hsv = cv.cvtColor(img_rgb, cv.COLOR_RGB2HSV)

# 调亮
lower_brightness = np.array([0, 50, 50])
upper_brightness = np.array([10, 255, 255])
bright_image = cv.inRange(img_hsv, lower_brightness, upper_brightness)

plt.figure(figsize=(12, 10))
plt.subplot(121), plt.imshow(img_rgb), plt.title('Original')
plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(cv.cvtColor(bright_image, cv.COLOR_BGR2RGB)), plt.title('bright_image')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# 将图像转换为灰度图像
img_gray = cv.cvtColor(img_rgb, cv.COLOR_RGB2GRAY)
plt.figure(figsize=(6, 4))
plt.imshow(cv.cvtColor(img_gray, cv.COLOR_BGR2RGB))
plt.show()
```

```python
# 绘制直方图
plt.hist(img_gray.ravel(), 256, [0,256]); plt.show()
```

```python
# 图像均衡化
equ = cv.equalizeHist(img_gray)
equ_img = np.hstack((img_gray, equ))
plt.figure(figsize=(10, 8))
plt.imshow(cv.cvtColor(equ_img, cv.COLOR_BGR2RGB)), plt.title(' Original--Equalization')
plt.show()
```
